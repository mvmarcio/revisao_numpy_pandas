Inicialmente vamos criar um *data frame* fict√≠cio com informa√ßoes das receitas de 10 empresas ao longo de 6 meses. 

```python
# Criando o data frame fict√≠cio

import numpy as np

df = np.random.uniform(1520, 10540, 60).reshape(10, 6).round(2)
df

# Inicialmente foram criados 60 valores aleat√≥rios que seguem uma distribui√ß√£o de
# probabilidae uniforme cont√≠nua, com valor m√≠nimo de R$ 1520,0 e m√°ximo de R$ 10.540,0. 
# Posteriormente esses valores foram organizados 
# da seguinte forma: 10 linhas e 6 colunas por meio da fun√ß√£o reshape. Foram 
# fixadas duas casas decimais para melhor se parecerem com valores monet√°rios.
```

Agora vamos colocar nomes nas empresas e nas receitas. Em homenagem ao Chaves e Chapolin, vamos nomear as empresas como Chapolin, Chaves, Madruga, Clotilde, Barriga, Girafales, Kiko, Chiquinha, Nhonho e Florinda que est√£o listadas na Bolsa de Valores Nossa Vila. As receitas v√£o seguir uma ordem de R1, R2, (...), R10. 

```python
# Criando os objetos das empresas e receitas. Ser√£o objetos do tipo lista

empresas = ['Chapolin', 'Chaves', 'Madruga', 'Clotilde', 'Barriga', 'Girafales', 
            'Kiko', 'Chiquinha', 'Nhonho', 'Florinda']

receitas = ['R1', 'R2', 'R3', 'R4', 'R5', 'R6']

# Vamos visualizar nossos objetos:
print(empresas)
print(receitas)

print(type(empresas)) # imprime o tipo de objeto
print(type(receitas)) # imprime o tipo de objeto
```

Agora vamos criar o nosso *data frame* utilizando as receitas como colunas e as empresas como linhas por meio da fun√ß√£o $pd.DataFrame(X, columns = y, index = w)$, em que $X$ √© a matriz com dos dados, $columns$ √© o vetor coluna e $index$ √© o vetor linha. 

```python
# Criando o data frame

resultados = pd.DataFrame(df, columns = receitas, index = empresas)
resultados
```

Tamb√©m poder√≠amos gerar o *data frame* tamb√©m com o uso de dicion√°rios. Vamos fazer essa opera√ß√£o a seguir para gerar as receitas das empresas. 

```python
# Criando o data frame por meio de dicion√°rios

dicionario = {'Chapolin': np.random.uniform(1520, 10450, 6).round(2),
              'Chaves': np.random.uniform(1520, 11000, 6).round(2),
              'Madruga': np.random.uniform(5000, 15000, 6).round(2),
              'Clotilde': np.random.uniform(1400, 9800, 6).round(2),
              'Barriga': np.random.uniform(1200, 7800, 6).round(2),
              'Girafales': np.random.uniform(1100, 6500, 6).round(2),
              'Kiko': np.random.uniform(900, 5400, 6).round(2),
              'Chiquinha': np.random.uniform(1400, 8000, 6).round(2),
              'Nhonho': np.random.uniform(1000, 6400, 6).round(2),
              'Florinda': np.random.uniform(850, 4000, 6).round(2)}

dicionario
```

```python
# Vamos acrescentar ao data frame

resultados = pd.DataFrame(dicionario)
resultados
```

```python
# Vamos acrescentar os nomes das linhas por meio do atributo index

resultados.index = ['R1', 'R2', 'R3', 'R4', 'R5', 'R6']

resultados
```

```python
# Agora vamos transpor o data frame para o nome das empresas fiquem nas linhas
# Vamos utilizar a fun√ß√£o transpose()

resultados = resultados.transpose()
resultados
```

```python
# Agora vamos saber o tipo te cada vari√°vel no banco de dados por meio do m√©todo info()

resultados.info()
```

### **Estat√≠sticas Descritivas**

Agora vamos trabalhar com algumas estat√≠sticas descritivas para avaliar as vari√°veis do banco de dados. Vamos utilizar inicialmente o m√©todo $describe(percentiles=None, include=None, exclude=None, datetime_is_numeric=False)$ do Pandas, em que:

i. **percentiles:** S√£o os quartis, quintis e/ou os dicis que desejamos avaliar;

ii. **include:** Mostra quais os resultados que das vari√°veis queremos avaliar. Como nosso banco de dados √© pequeno, vamos mostrar de todas as vari√°veis;

iii. **exclude:** Omite quais os resultados que das vari√°veis queremos n√£o avaliar;

iv. **datetime_is_numeric:** Argumento para fixa√ß√£o da vari√°vel temporal em s√©ries temporais. N√£o iremos trabalhar com esse tipo de dado hoje (mas com certeza vou escrever sobre no futuro üòä.

Ela nos fornece algumas estat√≠sticas importantes. Posteriormente vamos utilizar algumas outras estat√≠sticas importantes que n√£o est√£o no escopo da fun√ß√£o. 

As estat√≠sticas existem para fornecer informa√ß√µes relevantes sobre os dados, ou seja, o mais importante n√£o √© o c√°lculo em si, mas sua interpreta√ß√£o. Outra informa√ß√£o importante √© que por conven√ß√£o recomenda-se utilizar as estat√≠sticas nas colunas e as vari√°veis nas linhas. 

```python
# Estat√≠sticas descrtivas com a fun√ß√£o describe

resultados.T.describe(percentiles = [.25, .50, .75]).round(2)
```

Agora vamos interpretar os resultados da tabela de estat√≠sticas descritivas. Foram observadas 6 per√≠odos de receitas para cada empresa. A maior receita m√©dia foi a da empresa Madruga, no valor de 10.144,00 reais e a de menor receita m√©dia foi a da Florinda, com 1.971,46 reais. 

A empresa que apresentou maior dispers√£o entre as receitas no per√≠odo foi a Chapolin, com desvio padr√£o de 3.235,99 reais. A que obteve menor dispers√£o foi a Florinda, com desvio padr√£o de 968,95 reais. A diferen√ßa entre o valor m√≠nimo e m√°ximo da empresa Chapolin foi bem alto, no valor de 8.217,79 reais indicando forte discrep√¢ncia e corroborando com o resultado da dispers√£o. 

Os quartis das receitas das empresas foram bem pr√≥ximos, at√© mesmo na empresa Chapolin. Vamos fazer analisar a dispers√£o nos dados por do c√°lculo do coeficiente de varia√ß√£o, da assimetria e da curtose. O coeficiente de varia√ß√£o √© a divis√£o do desvio padr√£o pela m√©dia e indica de forma intuitiva de se os dados est√£o dispersos ou n√£o. S√£o considerados dispersos caso o coeficiente de varia√ß√£o for maior que 0,3 (30%). A assimetria e curtose t√™m o mesmo objetivo, uma assimetria acima de zero indica que os dados est√£o acim√©tricos √† direita e abaixo de zero indicam que est√£o assim√©tricos √† esquerda.

J√° a curtose indica se os dados est√£o espa√ßados √† direita ou √† esquerda. Neste caso, quando a curtose for maior que 3 (ou zero) a distribui√ß√£o dos dados √© mais concentrada, quando √© menor que 3 (ou zero) ent√£o os dados s√£o mais achatados. O valor 3 √© tomado como base por que √© o valor da curtose da distribui√ß√£o Normal, que √© sim√©trica. 

**Essas medidas s√£o importantes para avaliar uma boa toamda de decis√£o, pois caso haja dispers√£o nos dados a m√©dia n√£o √© melhor indicador de centralidade, preferindo-se trabalhar com a mediana. Isso √© mais importante ainda quando se for trabalhar com an√°lise preditiva, pois uma baixa dispers√£o √© fundamental para uma an√°lise mais assertiva, impactando poss√≠veis decis√µes estrat√©gicas de neg√≥cio.**

Caso haja forte dispers√£o nos dados, vamos avaliar a mediana ao inv√©s da m√©dia. 

```python
# Vamos construir algumas novas vari√°veis:

resultados['RM'] = resultados.mean(axis = 1).round(2) # vamos criar a coluna com a receita m√©dia
resultados['RQ'] = resultados.median(axis = 1) # vamos criar a coluna a mediana da receita 
resultados['CV'] = ((resultados.std(axis = 1) / resultados.mean(axis = 1))*100).round(2) # vamos criar a coluna com o coeficiente de varia√ß√£o
resultados['CT'] = resultados.kurt(axis = 1).round(2) # vamos criar uma coluna com a curtose das receitas
resultados['AS'] = resultados.skew(axis = 1).round(2) # vamos criar uma coluna com assimetria das receitas

resultados
```

```python
# Agora vamos avaliar com base no coeficente de varia√ß√£o se as receitas est√£o dispersas ou n√£o
# Vamos criar uma fun√ß√£o auxiliar que se CV > 30 ser√° considerada dispersa

def aux(x):
  if x > 30:
    return 'DISPERSA'
  else:
    return 'HOMOG√äNEA'

resultados['AVAL1'] = resultados['CV'].apply(aux)
```

```python
# Agora vamos avaliar com base na assimetria se as receitas est√£o assim√©tricas ou n√£o
# Vamos criar uma fun√ß√£o auxiliar que se AS != 0 ser√° considerada assim√©trica

def aux2(y):
  if y != 0:
    return 'ASSIM√âTRICA'
  else:
    return 'SIM√âTRICA'

resultados['AVAL2'] = resultados['AS'].apply(aux2)
```

```python
# Agora vamos avaliar com base na curtose se as receitas est√£o assim√©tricas ou n√£o
# Vamos criar uma fun√ß√£o auxiliar que se CT == 0 ser√° considerada mesoc√∫rtica,
# CT > 0 ser√° leptoc√∫rtica e se CT < 0 ser√° platic√∫rtica

def aux3(w):
  if w == 0:
    return 'MESOC√öRTICA'
  elif w > 0:
    return 'LEPTOC√öRTICA'
  elif w < 0:
    return 'PLATIC√öRTICA'

resultados['AVAL3'] = resultados['CT'].apply(aux3)
```

```python
# Agora vamos visualizar o banco de dados com as novas vari√°veis

resultados
```

```python
# Vamos avaliar qual empresa teve maior dipers√£o de receitas pelo coeficiente de varia√ß√£o

resultados.sort_values('CV', ascending = False)
```

Chapolin obteve a maior dispers√£o de receita. Numa situa√ß√£o real isso poderia indicar sazonalidade nas receitas da empresa ou movimentos at√≠picos. A equipe de estrat√©gia de vendas poderia cirar mecanismos para que as receitas pudessem ser potencializadas nos momentos de baixas. 

Vamos avaliar quem teve as maiores m√©dias e medianas de receitas durante o per√≠odo e comparar os resultados.

```python
# Avaliando as maiores m√©dias

resultados.sort_values('RM', ascending = False)
```

```python
# Avaliando as maiores medianas

resultados.sort_values('RQ', ascending = False)
```

Os resultados foram similares, mas houve uma mudan√ßa interessante: a empresa Nhonho teve um resultado superior ao da empresa Chapolin quando se analisou a mediana. Essa √© uma quest√£o muito interessante a ser discutida dentro de um neg√≥cio: 

1. √â economicamente vi√°vel operar nos meses em que h√° baixa receita?
